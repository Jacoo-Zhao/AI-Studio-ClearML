#  ClearML Pipeline Demo

This notebook demonstrates how to build and run a **task-based pipeline** using [ClearML](https://clear.ml), covering:

- Upload dataset
- Preprocess data
- Train a model
- Run a full pipeline


## Step 0: Initialize ClearML in SageMaker (via Terminal)

1. Open **Terminal** in JupyterLab

2. Install ClearML:

```bash
pip install -U clearml clearml-agent
```

3. Run initialization:

```bash
clearml-init
```

4. Follow prompts:

- Go to: https://app.clear.ml/settings/workspace-configuration
- Click **Create new credentials**
- Copy the configuration
- Paste into terminal when prompted

5. After completion, config will be saved to:

```bash
~/.clearml.conf
```
